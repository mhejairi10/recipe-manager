```python
# ============================================================
# main.py - Terminal Application
#
# Group Members:
# - Elias Abbas
# - Ali Radhi
# - Mohammed Hejairi
# ============================================================

import recipe_manager as rm
import pandas as pd
from datetime import date

CSV_FILE = "data/recipe.csv"


def add_new_recipe():
    print("\nAdd New Recipe")

    name = input("Recipe Name: ")
    ingredients = input("Ingredients separated by commas: ")
    time = (input("Preparation Time in Minutes: "))
    instructions = input("Cooking Instructions: ")
    difficulty = input("Difficulty (Easy / Medium / Hard): ")
    category = input("Category (Breakfast / Lunch / Dinner / Dessert): ")
    servings = (input("Number of Servings: "))
    rating = (input("Rating from 1 to 5: "))
    cooked = input("Have you cooked it before? (Yes/No/Today): ")


    if not name or not ingredients or not instructions:
        print("Please fill all required fields.")
        return

    if not rm.isValidInputName(name):
        print("Recipe name must contain only letters and spaces.")
        return

    if not rm.isValidInputIngredient(ingredients):
        print("Ingredients must contain only letters separated by commas.")
        return

    if not rm.isValidInputNumber(time):
        print("Time must be a number.")
        return

    if not rm.isValidInputNumber(servings):
        print("Servings must be a number.")
        return

    if not rm.isValidInputNumber(rating) or int(rating) < 1 or int(rating) > 5:
        print("Rating must be a number from 1 to 5.")
        return

    if not rm.isValidDifficulty(difficulty):
        print("Invalid difficulty. Choose Easy, Medium, or Hard.")
        return

    if not rm.isValidCategory(category):
        print("Invalid category. Choose Breakfast, Lunch, Dinner, or Dessert.")
        return

    if not rm.isValidCookedChoice(cooked):
        print("Invalid choice. Choose Yes, No, or Today.")
        return

    if cooked == "Yes":
        cooked_date = input("Enter cooked date (YYYY-MM-DD): ")

        if not rm.isValidDate(cooked_date):
            print("Invalid date. Use this format: YYYY-MM-DD")
            return

    elif cooked == "Today":
        cooked_date = date.today().strftime("%Y-%m-%d")

    else:
        cooked_date = None

    rm.save_recipe(
        name,
        ingredients,
        int(time),
        instructions,
        difficulty,
        category,
        int(servings),
        int(rating),
        cooked_date
    )

    print("Recipe added successfully!")

    
def search_by_ingredient():
    print("\nSearch Recipes by Ingredient")

    ingredient = input("Enter ingredient: ")

    if not rm.isValidInputName(ingredient):
        print("Ingredient must contain only letters.")
    else:
        results = rm.search(ingredient)

        if not results.empty:
            display(results.style.hide(axis='index'))
        else:
            print("No recipes found.")


def view_all_recipes():
    print("\nAll Recipes")

    results = rm.view_all()

    if results is not None:
        display(results.style.hide(axis='index'))
    else:
        print("Sorry, no recipes found.")


def random_recipe_suggestion():
    print("\nRandom Recipe Suggestion")

    results = rm.random_select()

    if results is not None:
        display(results.style.hide(axis='index'))
    else:
        print("Sorry, no recipes found.")


def generate_shopping_list():
    print("\nGenerate Shopping List")

    try:
        df = pd.read_csv(CSV_FILE)

        print("\nAvailable Recipes:")
        for recipe in df["Recipe"]:
            print("- " + recipe)

        selected = input("\nEnter recipe names separated by commas: ")
        if not rm.isValidInputIngredient(selected):
            print("Ingredients must contain only letters and spaces seperated by commas.")
            return
        selected_recipes = [item.strip() for item in selected.split(",")]

        shopping_list = rm.generate_shopping_list(selected_recipes)

        print("\nShopping List:")
        for item in shopping_list:
            print("- " + item)

    except FileNotFoundError:
        print("No recipe file found.")


def cooking_history():
    print("\nCooking History")

    results = rm.view_cooking_history()

    if results is not None:
        print(results.to_string(index=False))
    else:
        print("No cooking history found.")

    print("\nSuggestion")

    suggest = rm.suggest_by_history()

    if suggest is not None and not suggest.empty:
        print(suggest.to_string(index=False))

        recipe_name = suggest["Recipe"].iloc[0]
        choice = input("\nWill you cook this today? (Yes/No): ")

        if choice.lower() == "yes":
            rm.update_cooked_date(recipe_name)
            print("Cooking date updated!")
    else:
        print("No suggestion found.")

def sort_rating():
    

def main():
    while True:
        print("\n==============================")
        print("Recipe Manager")
        print("==============================")
        print("1. Add New Recipe")
        print("2. Search by Ingredient")
        print("3. View All Recipes")
        print("4. Random Recipe Suggestion")
        print("5. Generate Shopping List")
        print("6. Cooking History")
        print("7. Exit")

        choice = input("Choose an option: ")

        if choice == "1":
            add_new_recipe()
        elif choice == "2":
            search_by_ingredient()
        elif choice == "3":
            view_all_recipes()
        elif choice == "4":
            random_recipe_suggestion()
        elif choice == "5":
            generate_shopping_list()
        elif choice == "6":
            cooking_history()
        elif choice == "7":
            rm.sort_rating()
        elif choice == "8":
            print("Goodbye!")
            break
        
            
        else:
            print("Invalid choice. Please try again.")


```
