```python
import recipe_manager as rm

def main():
    #commnad line interface 
    while True:
        print("1. Add New Recipe")
        print("2. Search by Ingredient")
        print("3. View All Recipes")
        print("4. Random Recipe")
        print("5. Generate Shopping List")
        print("6. Cooking History")
        print("7. Exit")

        choice = input("Choose an option: ")

        if choice == "1":

            name = input("Recipe Name: ")
            ingredients = input("Ingredients (comma separated): ")
            time = int(input("Preparation Time: "))
            instructions = input("Instructions: ")
            difficulty = input("Difficulty (Easy/Medium/Hard): ")
            category = input("Category: ")
            servings = int(input("Servings: "))
            rating = int(input("Rating (1-5): "))

            cooked = input("Have you cooked it before? (yes/no): ")

            if cooked.lower() == "yes":
                cooked_date = input("Enter cooked date (YYYY-MM-DD): ")
            else:
                cooked_date = None

            rm.save_recipe(
                name,
                ingredients,
                time,
                instructions,
                difficulty,
                category,
                servings,
                rating,
                cooked_date
            )

        elif choice == "2":

            ingredient = input("Ingredient: ")
            print(rm.search(ingredient))

        elif choice == "3":

            print(rm.view_all())

        elif choice == "4":

            print(rm.random_select())

        elif choice == "5":

            recipes = input("Enter recipe names separated by commas: ")
            recipes = recipes.split(",")

            shopping = rm.generate_shopping_list(recipes)

            print("\nShopping List")

            for item in shopping:
                print("-", item)

        elif choice == "6":

            print("\nCooking History")
            print(rm.view_cooking_history())

            print("\nSuggested Recipe")
            suggestion = rm.suggest_by_history()
            print(suggestion)

        elif choice == "7":

            print("Goodbye!")
            break

        else:

            print("Invalid choice.")
```

```python
main()
```
